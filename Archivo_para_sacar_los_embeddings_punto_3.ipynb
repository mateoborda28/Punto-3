```python
import stanza
import spacy
from sklearn.preprocessing import StandardScaler


def evaluacion(train1, test1, columna, metodo, ingenieria):
    train1["muestra"] = "Entrenamiento"
    test1["muestra"] = "Test"
    datos = pd.concat([train1, test1], axis = 0)
    datos['Sarcasmo'] = datos['Sarcasmo'].apply(lambda x: 1 if x == 'Si' else 0)
    tool = language_tool_python.LanguageTool('es')
    # Verificar si la columna existe
    if columna not in datos.columns:
        raise ValueError(f"La columna '{columna}' no existe en el DataFrame.")

    # Cambiar el nombre de la columna a 'texto_procesado'
    datos = datos.rename(columns={columna: 'texto_procesado'})

    # Aplicar corrección gramatical con manejo de errores
    def analizar_texto(txt):
        try:
            matches = tool.check(txt)
            corrected = language_tool_python.utils.correct(txt, matches)
            return matches, len(matches), corrected
        except Exception as e:
            print(f"Error al procesar el texto: {txt}. Error: {e}")
            return [], 0, txt


    # Crear nuevas columnas con la evaluación gramatical
    datos[['matches', 'corrections', 'corrected']] = datos['texto_procesado'].apply(
        lambda txt: pd.Series(analizar_texto(txt))
    )

    sents = [] # formar listas con las oraciones de cada texto
    tokens = [] # separar las palabras y caracteres de cada texto
    lemma = [] # convertir las palabras en su raíz (ayuda --> ayudar)
    pos = [] # indica en cada posición si la palabra es pronombre, conjunción, etc
    ner = [] # indica si la palabra es considerada una entidad (nombre propio importante en el texto)

    stop_words = set(STOP_WORDS)
    # stop_words.update(punctuation) # remove it if you need punctuation
    nlp = spacy.load('es_core_news_lg') ## ESPAÑOL
    for essay in nlp.pipe(datos['corrected'], batch_size=100, n_process=4): ### el 100 es para que3 vaya de 100 en 100 ##
     if essay.is_parsed:
        tokens.append([e.text for e in essay])
        sents.append([sent.text.strip() for sent in essay.sents])
        pos.append([e.pos_ for e in essay])
        ner.append([e.text for e in essay.ents])
        lemma.append([n.lemma_ for n in essay])
     else:
        tokens.append(None)
        lemma.append(None)
        pos.append(None)
        sents.append(None)
        ner.append(None)


    datos['tokens'] = tokens
    datos['lemma'] = lemma
    datos['pos'] = pos
    datos['sents'] = sents
    datos['ner'] = ner

    datos['token_count'] = datos.apply(lambda x: len(x['tokens']), axis=1)
    datos['unique_token_count'] = datos.apply(lambda x: len(set(x['tokens'])), axis=1)
    datos['nostop_count'] = datos.apply(lambda x: len([token for token in x['tokens'] if token not in stop_words]), axis=1)
    datos['sent_count'] = datos.apply(lambda x: len(x['sents']), axis=1)
    datos['ner_count'] = datos.apply(lambda x: len(x['ner']), axis=1)
    datos['comma'] = datos.apply(lambda x: x['corrected'].count(','), axis=1)
    datos['question'] = datos.apply(lambda x: x['corrected'].count('?'), axis=1)
    datos['exclamation'] = datos.apply(lambda x: x['corrected'].count('!'), axis=1)
    datos['quotation'] = datos.apply(lambda x: x['corrected'].count('"') + x['corrected'].count("'"), axis=1)
    datos['organization'] = datos.apply(lambda x: x['corrected'].count(r'@ORGANIZATION'), axis=1)
    datos['caps'] = datos.apply(lambda x: x['corrected'].count(r'@CAPS'), axis=1)
    datos['person'] = datos.apply(lambda x: x['corrected'].count(r'@PERSON'), axis=1)
    datos['location'] = datos.apply(lambda x: x['corrected'].count(r'@LOCATION'), axis=1)
    datos['money'] = datos.apply(lambda x: x['corrected'].count(r'@MONEY'), axis=1)
    datos['time'] = datos.apply(lambda x: x['corrected'].count(r'@TIME'), axis=1)
    datos['date'] = datos.apply(lambda x: x['corrected'].count(r'@DATE'), axis=1)
    datos['percent'] = datos.apply(lambda x: x['corrected'].count(r'@PERCENT'), axis=1)
    datos['noun'] = datos.apply(lambda x: x['pos'].count('NOUN'), axis=1)
    datos['adj'] = datos.apply(lambda x: x['pos'].count('ADJ'), axis=1)
    datos['pron'] = datos.apply(lambda x: x['pos'].count('PRON'), axis=1)
    datos['verb'] = datos.apply(lambda x: x['pos'].count('VERB'), axis=1)
    datos['noun'] = datos.apply(lambda x: x['pos'].count('NOUN'), axis=1)
    datos['cconj'] = datos.apply(lambda x: x['pos'].count('CCONJ'), axis=1)
    datos['adv'] = datos.apply(lambda x: x['pos'].count('ADV'), axis=1)
    datos['det'] = datos.apply(lambda x: x['pos'].count('DET'), axis=1)
    datos['propn'] = datos.apply(lambda x: x['pos'].count('PROPN'), axis=1)
    datos['num'] = datos.apply(lambda x: x['pos'].count('NUM'), axis=1)
    datos['part'] = datos.apply(lambda x: x['pos'].count('PART'), axis=1)
    datos['intj'] = datos.apply(lambda x: x['pos'].count('INTJ'), axis=1)
    datos['corrected'].to_csv(path + 'NLP_texto_corregido.csv')

    categorias_gramaticales = datos.get(["corrected","tokens",
           "corrections","token_count","unique_token_count",
           "nostop_count","sent_count","ner_count","comma","question",
           "exclamation", "quotation", "organization", "caps", "person",
           "location", "money", "time", "date", "percent", "noun", "adj", "pron",
           "verb", "cconj", "adv", "det", "propn", "num", "part", "intj"])

    # Cargar el modelo BERT preentrenado para español
    tokenizer = BertTokenizer.from_pretrained('dccuchile/bert-base-spanish-wwm-cased')
    model = BertModel.from_pretrained('dccuchile/bert-base-spanish-wwm-cased')

    ### Procesamiento de texto
    # Aplicamos el procesamiento y lematización que ya tenías
    datos['processed_text'] = datos.apply(lambda row: ' '.join(token.lemma_ for token in nlp(row["corrected"]).sents), axis=1)
    datos['processed_text'] = datos['processed_text'].str.lower()
    datos['processed_text'] = datos['processed_text'].replace(list('áéíóú'), list('aeiou'), regex=True)
    datos['processed_text'] = datos['processed_text'].str.replace('[^\w\s]', '')
    # Unir todos los textos procesados en una sola cadena para contar las frecuencias
    all_text = ' '.join(datos['processed_text'])

    # Tokenizar y contar la frecuencia de cada palabra
    word_counts = Counter(all_text.split())
    # Definir un umbral de frecuencia, por ejemplo, palabras que aparecen más de 5 veces
    threshold = 5
    common_words = {word for word, count in word_counts.items() if count > threshold}
    # Filtrar palabras que tienen más de 4 caracteres y están en el conjunto de palabras frecuentes
    datos['processed_text'] = datos['processed_text'].apply(lambda x: ' '.join([word for word in x.split() if len(word) > 4 and word in common_words]))
    numericas = datos.select_dtypes(include=['float64', 'int64']).columns
    trainc = datos[datos["muestra"] == "Entrenamiento"].drop(columns=["muestra"])
    testc = datos[datos["muestra"] == "Test"].drop(columns=["muestra"])
#### Ingenieria de variables ########
    # Variables al cuadrado
    categorias_gramaticales_numericas = [var for var in categorias_gramaticales if var in numericas]
    train_con_ingenieria_b_varcuadrado = pd.DataFrame()
    test_con_ingenieria_b_varcuadrado = pd.DataFrame()
    # Elevar al cuadrado solo las columnas numéricas
    for var in categorias_gramaticales_numericas:
     train_con_ingenieria_b_varcuadrado[f'x_{var}_2'] = trainc[var] ** 2
    for var in categorias_gramaticales_numericas:
     test_con_ingenieria_b_varcuadrado[f'x_{var}_2'] = testc[var] ** 2
    cov = list(train_con_ingenieria_b_varcuadrado.columns)
    cov1 = list(test_con_ingenieria_b_varcuadrado.columns)
    X1 = train_con_ingenieria_b_varcuadrado.get(cov)
    x2 = test_con_ingenieria_b_varcuadrado.get(cov1)
    y1 = trainc.get(["Sarcasmo"])
    y2 = testc.get(["Sarcasmo"])
    modelo1 = XGBClassifier()
    modelo1 = modelo1.fit(X1,y1)
    modelo2 = XGBClassifier()
    modelo2 = modelo2.fit(x2,y2)
    importancias = modelo1.feature_importances_
    importancias2 = modelo2.feature_importances_
    imp1 = pd.DataFrame({'Variable':X1.columns,'Importancia':importancias})
    imp2 = pd.DataFrame({'Variable':x2.columns,'Importancia':importancias2})
    imp1["Importancia"] = imp1["Importancia"] * 100 / np.sum(imp1["Importancia"])
    imp2["Importancia"] = imp2["Importancia"] * 100 / np.sum(imp2["Importancia"])
    imp1 = imp1.sort_values(["Importancia"],ascending=False)
    imp2 = imp2.sort_values(["Importancia"],ascending=False)
    imp1.index = range(imp1.shape[0])
    imp2.index = range(imp2.shape[0])
    var_al_cuadrado_seleccionadas = imp1.head(5)["Variable"].values
    var_al_cuadrado_seleccionadas2 = imp2.head(5)["Variable"].values
    var_al_cuadrado_seleccionadas_df = pd.DataFrame(var_al_cuadrado_seleccionadas, columns=['Variable'])
    var_al_cuadrado_seleccionadas2_df = pd.DataFrame(var_al_cuadrado_seleccionadas2, columns=['Variable'])
    ### Interacciones
    train_interacciones = pd.DataFrame()
    test_interacciones = pd.DataFrame()
    if not categorias_gramaticales_numericas:
     print("Error: No hay columnas numéricas en las categorías gramaticales.")
    else:
    # Crear interacciones entre las columnas numéricas
     for i in range(len(categorias_gramaticales_numericas)):
        for j in range(i + 1, len(categorias_gramaticales_numericas)):
            var1 = categorias_gramaticales_numericas[i]
            var2 = categorias_gramaticales_numericas[j]
            # Crear el término de interacción con nombres descriptivos
            new_column_name = f'x_{var1}_x_{var2}'
            train_interacciones[new_column_name] = trainc[var1] * trainc[var2]
            new_column_name2 = f'x_{var1}_x_{var2}'
            test_interacciones[new_column_name2] = testc[var1] * testc[var2]
    cov3 = list(train_interacciones.columns)
    cov4 = list(test_interacciones.columns)
    X3 = train_interacciones.get(cov3)
    x4 = test_interacciones.get(cov4)
    y3 = trainc.get(["Sarcasmo"])
    y4 = testc.get(["Sarcasmo"])
    modelo3 = XGBClassifier()
    modelo3 = modelo3.fit(X3,y3)
    modelo4 = XGBClassifier()
    modelo4 = modelo4.fit(x4,y4)
    importancias3 = modelo3.feature_importances_
    importancias4 = modelo4.feature_importances_
    imp3 = pd.DataFrame({'Variable':X3.columns,'Importancia':importancias3})
    imp4 = pd.DataFrame({'Variable':x4.columns,'Importancia':importancias4})
    imp3["Importancia"] = imp3["Importancia"] * 100 / np.sum(imp3["Importancia"])
    imp4["Importancia"] = imp4["Importancia"] * 100 / np.sum(imp4["Importancia"])
    imp3 = imp3.sort_values(["Importancia"],ascending=False)
    imp4 = imp4.sort_values(["Importancia"],ascending=False)
    imp3.index = range(imp3.shape[0])
    imp4.index = range(imp4.shape[0])
    interacciones_seleccionadas = imp3.head(5)["Variable"].values
    interacciones_seleccionadas2 = imp4.head(5)["Variable"].values
    interacciones_seleccionadas_df = pd.DataFrame(interacciones_seleccionadas, columns=['Variable'])
    interacciones_seleccionadas2_df = pd.DataFrame(interacciones_seleccionadas2, columns=['Variable'])
    ### Razones
    train_razones = pd.DataFrame()
    test_razones = pd.DataFrame()
    if not categorias_gramaticales_numericas:
     print("Error: No hay columnas numéricas en las categorías gramaticales.")
    else:
    # Crear razones entre las columnas numéricas
     for i in range(len(categorias_gramaticales_numericas)):
        for j in range(i + 1, len(categorias_gramaticales_numericas)):
            var1 = categorias_gramaticales_numericas[i]
            var2 = categorias_gramaticales_numericas[j]
            # Evitar división por cero
            with np.errstate(divide='ignore', invalid='ignore'):
                # Crear el término de razón (var1 / var2)
                ratio_column_name_1 = f'x_{var1}_div_{var2}'
                train_razones[ratio_column_name_1] = trainc[var1] / (trainc[var2]+0.01)
                test_razones[ratio_column_name_1] = testc[var1] / (testc[var2]+0.01)
                # Crear el término de razón inversa (var2 / var1)
                ratio_column_name_2 = f'x_{var2}_div_{var1}'
                train_razones[ratio_column_name_2] = trainc[var2] / (trainc[var1]+0.01)
                test_razones[ratio_column_name_2] = testc[var2] / (testc[var1]+0.01)
    cov5 = list(train_razones.columns)
    cov6 = list(test_razones.columns)
    X5 = train_razones.get(cov5)
    x6 = test_razones.get(cov6)
    y5 = trainc.get(["Sarcasmo"])
    y6 = testc.get(["Sarcasmo"])
    modelo5 = XGBClassifier()
    modelo5 = modelo5.fit(X5,y5)
    modelo6 = XGBClassifier()
    modelo6 = modelo6.fit(x6,y6)
    importancias5 = modelo5.feature_importances_
    importancias6 = modelo6.feature_importances_
    imp5 = pd.DataFrame({'Variable':X5.columns,'Importancia':importancias5})
    imp6 = pd.DataFrame({'Variable':x6.columns,'Importancia':importancias6})
    imp5["Importancia"] = imp5["Importancia"] * 100 / np.sum(imp5["Importancia"]+0.01)
    imp6["Importancia"] = imp6["Importancia"] * 100 / np.sum(imp6["Importancia"]+0.01)
    imp5 = imp5.sort_values(["Importancia"],ascending=False)
    imp6 = imp6.sort_values(["Importancia"],ascending=False)
    imp5.index = range(imp5.shape[0])
    imp6.index = range(imp6.shape[0])
    razones_seleccionadas = imp5.head(5)["Variable"].values
    razones_seleccionadas2 = imp6.head(5)["Variable"].values
    razones_seleccionadas_df = pd.DataFrame(razones_seleccionadas, columns=['Variable'])
    razones_seleccionadas2_df = pd.DataFrame(razones_seleccionadas2, columns=['Variable'])
    ingenieria_de_variables_train = pd.concat([trainc, var_al_cuadrado_seleccionadas_df, interacciones_seleccionadas_df, razones_seleccionadas_df], axis=1)
    ingenieria_de_variables_test = pd.concat([testc, var_al_cuadrado_seleccionadas2_df, interacciones_seleccionadas2_df, razones_seleccionadas2_df], axis=1)

    #### Embeddings con Bert
    def get_bert_embedding(text):
     inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512)
     outputs = model(**inputs)
     # Usar la representación de la [CLS] token
     cls_embedding = outputs.last_hidden_state[:, 0, :].detach().numpy()
     return cls_embedding.flatten()
    # Aplicar BERT a cada texto
    trainc['bert_embedding'] = trainc['processed_text'].apply(get_bert_embedding)
    testc['bert_embedding'] = testc['processed_text'].apply(get_bert_embedding)
    # Convertir embeddings a una matriz para PCA y clustering
    embeddings_train = np.stack(trainc['bert_embedding'].values)
    embeddings_test = np.stack(testc['bert_embedding'].values)
    # Reducción de dimensionalidad con PCA para visualizar
    pca = PCA(n_components=2)
    embeddings_pca = pca.fit_transform(embeddings_train)
    embeddings_pca2 = pca.fit_transform(embeddings_test)
    # Crear un DataFrame con las proyecciones PCA
    df_pca = pd.DataFrame(embeddings_pca, columns=['x', 'y'])
    df_pca2 = pd.DataFrame(embeddings_pca2, columns=['x', 'y'])
    embedding_train_df = pd.DataFrame(embeddings_train)
    embedding_test_df = pd.DataFrame(embeddings_test)

    texto_numpy = np.array(trainc['processed_text'])
    texto_numpy2 = np.array(testc['processed_text'])

# Guardar el array como un archivo de texto
    with open('temp_text.txt', 'w') as f:
     for texto in texto_numpy:
        f.write(texto + "\n")

    with open('temp_text2.txt', 'w') as f:
     for texto in texto_numpy2:
        f.write(texto + "\n")


#### Fast Text ##################
    ft_model = fasttext.train_unsupervised('temp_text.txt',dim=300)
    ft_model2 = fasttext.train_unsupervised('temp_text2.txt',dim=300)

    no_doc = np.random.randint(len(trainc))
    no_doc2 = np.random.randint(len(testc))
    vocabulary = set(ft_model.words)

    def doc_word_vectors(words, model, vocabulary, num_features):
       nwords = 0.
       no_found=[]
       doc=[]
       for word in words:
         if word in vocabulary:
            nwords = nwords + 1.
            doc.append(model.get_word_vector(word))
         else:
            no_found.append(word)
       return np.array(doc),no_found

    emb0,no_found = doc_word_vectors(trainc['tokens'].iloc[no_doc], model=ft_model, vocabulary=vocabulary, num_features=ft_model.dim)
    emb_test, no_found_test = doc_word_vectors(testc['tokens'].iloc[no_doc2], model=ft_model, vocabulary=vocabulary, num_features=ft_model.dim)

    def average_word_vectors(words, model, vocabulary, num_features):

     feature_vector = np.zeros((num_features,),dtype="float64")
     nwords = 0.

     for word in words:
        if word in vocabulary:
            nwords = nwords + 1.
            feature_vector = np.add(feature_vector, model.get_word_vector(word))
     if nwords:
        feature_vector = np.divide(feature_vector, nwords)
     return feature_vector
    def averaged_word_vectorizer(corpus, model, num_features):
      vocabulary = set(model.words)
      features = [average_word_vectors(tokenized_sentence, model, vocabulary, num_features)
                    for tokenized_sentence in corpus]
      return np.array(features)

    ftext_feature_array_train = averaged_word_vectorizer(corpus=trainc['tokens'], model=ft_model, num_features=ft_model.dim)
    ftext_feature_array_test = averaged_word_vectorizer(corpus=testc['tokens'], model=ft_model, num_features=ft_model.dim)
    embedding_fast_train = pd.DataFrame(ftext_feature_array_train)
    embedding_fast_test = pd.DataFrame(ftext_feature_array_test)
#### Bien ##########
#### Selección del modelo ###

    embedding_vector_length = 768 if metodo == 'bert' else 300


    embeddings_train = embedding_train_df.copy()
    df_train = ingenieria_de_variables_train if ingenieria.lower() == 'si' else trainc
    features = df_train.columns[df_train.dtypes != 'object'].tolist()
    features_train = df_train.get(features)
    y_train = trainc.get(["Sarcasmo"]).to_numpy()

    # Para el conjunto de prueba
    embeddings_test =  embedding_test_df.copy()
    df_test = ingenieria_de_variables_test if ingenieria.lower() == 'si' else testc
    features1 = df_test.columns[df_test.dtypes != 'object'].tolist()
    features_test = df_test.get(features1)
    y_test = testc.get(["Sarcasmo"]).to_numpy()

    features_train.columns = features_train.columns.astype(str)
    features_test.columns = features_test.columns.astype(str)

    scaler = MinMaxScaler((-1.0,1.0))
    features_train_scaled = pd.DataFrame(scaler.fit_transform(features_train))
    features_train_scaled.columns = features_train.columns
    features_test_scaled = pd.DataFrame(scaler.fit_transform(features_test))
    features_test_scaled.columns = features_test.columns
##### bien #####

    embedding_vector = 768 if metodo == 'bert' else 300

    x1 = Input(shape=(embedding_vector,), name='Input_Embedding')
    x2 = Input(shape=(features_train_scaled.shape[1],), name='Input_Features')
    # Capa entrada
    x = Concatenate(name='Concatenar')([x1, x2])
    x = Dropout(0.50)(x)
    # capas ocultas
    x = Dense(64, activation='elu', name='Capa_Densa_1')(x)
    x = Dropout(0.25)(x)
    x = Dense(32, activation='elu', name='Capa_Densa_2')(x)
    x = Dropout(0.25)(x)
    x = Dense(16, activation='elu', name='Capa_Densa_3')(x)
    x = Dropout(0.25)(x)
    # Capa de salida para clasificación binaria
    x = Dense(1, activation='sigmoid', name='Output')(x)
    model = Model(inputs=[x1, x2], outputs=x)
    # Compilación para clasificación binaria
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    history = model.fit(
    x=[embedding_train_df, features_train_scaled],  # Entrenamos con embeddings y features escaladas
    y=y_train,
    validation_data=([embedding_test_df, features_test], y_test),  # Validación con datos de prueba
    epochs=100,
    batch_size=32,
    verbose=1)

    #Predicciones del train
    y_pred = model.predict([embeddings_train,features_train_scaled])
    y_true = y_train
    y_pred = (y_pred >= 0.5).astype(int)
    np.sum(y_pred)
    #Evaluación del train
    accuracy_score(y_true,y_pred)
    roc_auc_score(y_true,y_pred)
    #Predicciones del test
    y_pred = model.predict([embeddings_test,features_test_scaled])
    y_true = y_test
    y_pred = (y_pred>=0.5).astype(int)
    accuracy = accuracy_score(y_true,y_pred)
    roc = roc_auc_score(y_true,y_pred)
    embedding_fast_entrenamiento=embedding_fast_train.to_csv(path + 'embedding_fast_train.csv', index=False)
    embedding_fast_prueba = embedding_fast_test.to_csv(path + 'embedding_fast_test.csv', index=False)
    embedding_bert_entrenamiento = embedding_train_df.to_csv(path + 'embedding_train_df.csv', index=False)
    embedding_bert_prueba = embedding_test_df.to_csv(path + 'embedding_test_df.csv', index=False)
    ingenieria_entrenamiento = ingenieria_de_variables_train.to_csv(path + 'ingenieria_de_variables_train.csv', index=False)
    ingenieria_prueba = ingenieria_de_variables_test.to_csv(path + 'ingenieria_de_variables_test.csv', index=False)
    entrenamiento_sin_ingenieria = trainc.to_csv(path + 'entrenamiento_sin_ingenieria.csv', index=False)
    prueba_sin_ingenieria = testc.to_csv(path + 'prueba_sin_ingenieria.csv', index=False)


    return embedding_fast_entrenamiento, embedding_fast_prueba, embedding_bert_entrenamiento, embedding_bert_prueba, ingenieria_entrenamiento, ingenieria_prueba, entrenamiento_sin_ingenieria, prueba_sin_ingenieria, accuracy, roc
```
